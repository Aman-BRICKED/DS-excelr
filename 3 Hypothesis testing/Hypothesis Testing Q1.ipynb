```python
#Q1
#     A F&B manager wants to determine whether there is any significant difference
#in the diameter of the cutlet between two units. A randomly selected sample of cutlets
#was collected from both units and measured? Analyze the data and draw inferences at 5% significance level.
#Please state the assumptions and tests that you carried out to check validity of the assumptions.


    # Minitab File : Cutlets.mtw
```

```python
#HO : no difference between the diameter i.e U1 = U2
#H1 : significant difference between the diameter i.e U1 != U2
# we will use 2 sample t-test with significance level 95% i.e Alpha = 0.95
```

```python
import pandas as pd
import statsmodels.api as sm
import scipy.stats as stats
from scipy.stats import norm
```

```python
Q1 = pd.read_csv('Cutlets.csv')
Q1
Q1.rename(columns = {'Unit A' : 'Unit_1', 'Unit B' : 'Unit_2'}, inplace = True)
```

```python
#first check if the data is normally distributed to check we use Q-Q plot
sm.qqplot(Q1.Unit_1, line ='s')
sm.qqplot(Q1.Unit_2, line ='s')
#both appear to be normally dist.
```

```python
mean1 = Q1['Unit_1'].mean()
mean2 = Q1['Unit_2'].mean()
print(mean1 ,mean2)
```

```python
#carrying out two sample t test
stats.ttest_ind(Q1.Unit_1, Q1.Unit_2) 
#p value is 0.47 which is greater than 0.05. so cannot reject null hypothesis.
#no significance difference between diameter of cutlets of two units.
```
